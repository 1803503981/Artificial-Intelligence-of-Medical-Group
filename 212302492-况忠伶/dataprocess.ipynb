## 2.将各辨证类型中特征的占比保存为csv

```python
#将各辨证类型中特征的占比保存为csv
df2=pd.DataFrame()
for label in ['脾胃虚弱','风热侵袭','痰火郁结','肝火上扰','肾精亏虚']:
    df1=pd.DataFrame()
    # print(label)
    data1=data[data["中医辨证"] == f"{label}"]
    features_name1=data1.columns
    names1=features_name1[1:-22]#获取发病节气-脉象的特征
    for item in names1:
        list1 = data1[f"{item}"].values.tolist()
        print(item,list1)
        list_unique=np.unique(list1, return_counts=True)
        print(list_unique)
        dict1 = dict(zip(list_unique[0],np.round(list_unique[1]/len(data1),6)))
        # df = pd.DataFrame.from_dict(dict1, orient='index')
        df = pd.DataFrame(list(dict1.items()),columns=['特征名', '特征占比'])
        df=df[~df['特征名'].isin(['0'])]
        df1 = df1.append(df)
        df1 = df1.reset_index(drop=True)
        print(df1)
    df1.columns = [f'{label}',f'{label}特征占比']
    df2=pd.concat([df1, df2],axis = 1)
```

```python
from sklearn.model_selection import KFold,StratifiedKFold
from sklearn.utils import shuffle

def getKFoldData(data,fileName):
    df=data.sample(frac=1)
    # df = shuffle(data)
    y = df["中医辨证"]
    #.iloc[行，列]，前面的:表示提取所有行，后面的为切片，提取倒数第一列
    classes=np.unique(y)
    # data_df = pd.DataFrame(data_np, columns=columns)
    # print(data_df)
    # kfold = KFold(n_splits=5,random_state=666,shuffle=True)
    kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=999)
    split_i=0
    for train_index, test_index  in kfold.split(df,y):
        split_i=split_i+1
        # print("test index: ", test_index)
        Train=data.iloc[train_index]
        Test=data.iloc[test_index]
        print("Train,Test",Train,Test)
        print(Train["中医辨证"].value_counts(),Test["中医辨证"].value_counts())
        Train.to_csv(f'./data/五折训练集{split_i}.csv')
        Test.to_csv(f'./data/五折测试集{split_i}.csv')
```
